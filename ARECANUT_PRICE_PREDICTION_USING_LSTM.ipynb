```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('/content/data (3).csv')
data
```

```python
data=data[data.variety=='Rashi']
data=data[data.market_name=='Shimoga']
```

```python
data
```

```python
min_price=data['min_price']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
min_price=scaler.fit_transform(np.array(min_price).reshape(-1,1))
```

```python
min_price.shape
```

```python
train_length=int(len(min_price)*0.65)
train_min_price=min_price[0:train_length]
test_min_price=min_price[train_length:len(min_price)]
```

```python
train_min_price.shape
```

```python
test_min_price.shape
```

```python
def splitter(data,mov_size):
  x_side=[]
  y_side=[]
  for i in range(len(data)-mov_size-1):
    x=train_min_price[i:(i+mov_size),0]
    y=data[i+mov_size,0]
    x_side.append(x)
    y_side.append(y)
  return np.array(x_side),np.array(y_side)
  # print(type(x))
```

```python
x_train,y_train=splitter(train_min_price,4)
x_test,y_test=splitter(test_min_price,4)
```

```python
x_train.shape
```

```python
import tensorflow
from tensorflow.keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
model=Sequential()
model.add(LSTM(20,return_sequences=True,input_shape=(4,1)))
model.add(LSTM(20,return_sequences=True))
model.add(LSTM(20))
model.add(Dense(1))
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
model.summary()
```

```python
x_train=x_train.reshape(x_train.shape[0],x_train.shape[1],1)
```

```python
x_test=x_test.reshape(x_test.shape[0],x_test.shape[1],1)
```

```python
type(y_test)
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=30,batch_size=2)
```

```python
train_predict=model.predict(x_train)
test_predict=model.predict(x_test)
```

```python
train_predict=scaler.inverse_transform(train_predict)
test_predict=scaler.inverse_transform(test_predict)
```

```python
import math
from sklearn.metrics import mean_squared_error
```

```python
math.sqrt(mean_squared_error(y_train,train_predict))
```

```python
math.sqrt(mean_squared_error(y_test,test_predict))
```

```python
look=2
trainingplot=np.empty_like(min_price)
trainingplot[:,:]=np.nan
trainingplot[look:len(train_predict)+look,:]=train_predict
```

```python
testingplot=np.empty_like(min_price)
testingplot[:,:]=np.nan
testingplot[len(train_predict)+(look*2)+3:len(min_price)-3,:]=test_predict
```

```python
import matplotlib.pyplot as plt
plt.plot(scaler.inverse_transform(min_price))
plt.plot(trainingplot)
plt.plot(testingplot)
plt.show()
```
